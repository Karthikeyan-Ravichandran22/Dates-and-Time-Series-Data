# Working with Dates and Time Series Data

```python
import pandas as pd
```

```python
df=pd.read_csv("date.csv")
```

```python
df.head()
```

```python
df.loc[0,"Date"]
```

Formating Data Time According To data 

```python
df["Date"]=pd.to_datetime(df["Date"],format="%Y-%m-%d %I-%p",errors="raise")
```

```python
df["Date"]
```

Method for getting Day of Date

```python
df.loc[0,"Date"].day_name()
```

```python
df["Day_of_the_week"]=df["Date"].dt.day_name()
df["Day_of_the_week"]
```

```python
df["Date"].min()
```

```python
df["Date"].max()
```

```python
df["Date"].max()-df["Date"].min()
```

Here If we Try to Filter through day its not possiable hence in Upcomeing we convert date column as ''Index''

```python
filt=(df["Date"]>="2019") & (df["Date"]<="2020")
```

```python
df.loc[filt]
```

Here We covert Date column as Index so that we can work with date conviently

```python
df.set_index("Date",inplace=True)
```

```python
df.loc["2019-01":"2020-01"]
```

```python
df.loc["2019-01":"2020-01"].mean()
```

```python
df.loc["2019-08-22"]
```

```python
df.loc["2019-08-22"]["High"].max()
```

Using Resample we can convert date as per our request(eg: as Day,week,hour,month etc) as Fumction

```python
High_of_day=df["High"].resample("D").max()
High_of_day
```

```python
High_of_day.loc["2019-08-22"]
```

```python
mean_of_col=df.resample("W").mean()
mean_of_col
```

```python
df.resample("W").agg({"Close":"min","High":"max","Low":"min","Volume":"sum"})
```
